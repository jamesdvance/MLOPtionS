```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# scrape single image
test_page = "https://www.halloweencostumes.com/ghostbusters-mens-deluxe-costume.html"
```

```python
def get_soup(url):
    response = requests.get(url)
    return BeautifulSoup(response.text, 'html.parser')
```

```python
soup = get_soup(test_page)
```

```python
soup
```

```python
costume = soup.find(attrs={'property':'og:title'})['content'].strip()
print(costume)
```

```python
content_properties = ['og:image',"twitter:image", "og:video"]
```

```python
media = []
for cp in content_properties:
    media.append(soup.find(attrs={"property": cp})["content"])
```

```python
media
```

```python
review_links = soup.findAll(attrs={"class": "reviewGallery justify-content-center d-flex"})
review_imgs = []
for r in review_links:
    review_imgs.append(r.find("a")["href"])
```
